```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('BIKE DETAILS.csv')
```

```python
# Desviacin estandar
df['selling_price'].std()
```

```python
# Rango = valor max - valor min
df['selling_price'].max() - df['selling_price'].min()
```

```python
# Cuartiles
median = df['selling_price'].median()
q1 = df['selling_price'].quantile(q=0.25)
q3 = df['selling_price'].quantile(q=0.75)
min_val = df['selling_price'].quantile(q=0)
max_val = df['selling_price'].quantile(q=1.0)
print("Min=", min_val, "Q1=", q1, "Q2=", median, "Q3=", q3, "Max=", max_val)
```

```python
# Rango intercuartil
iqr = q3 - q1
iqr
```

```python
min_limit = q1 - (1.5 * iqr)
max_limit = q3 + (1.5 * iqr)
print("Valor mnimo =", min_limit, "| Valor mximo =", max_limit)
```
